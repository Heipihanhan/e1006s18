#### Turing Machine

There is a formal, mathematical definition of a Turing Machine.


Informally, a [Turing machine](https://en.wikipedia.org/wiki/Turing_machine) consists of:

- A tape divided into cells, one next to the other.
    - Each cell contains a symbol from some finite alphabet.
    - The alphabet contains a special blank symbol (here written as '0') and one or more other symbols.
    - The tape is assumed to be arbitrarily extendable to the left and to the right, i.e., the Turing machine is always supplied with as much tape as it needs for its computation.
    - Cells that have not been written before are assumed to be filled with the blank symbol.
    - In some models the tape has a left end marked with a special symbol; the tape extends or is indefinitely extensible to the right.
    
    
- A head that can read and write symbols on the tape and move the tape left and right one (and only one) cell at a time. In some models the head moves and the tape is stationary.


- A state register that stores the state of the Turing machine, one of finitely many.
    - Among these is the special start state with which the state register is initialized. 
    - These states, writes Turing, replace the "state of mind" a person performing computations would ordinarily be in.
    
    
- A finite table of instructions that, given the state(q<sub>i</sub>) the machine is currently in and the symbol(a<sub>j</sub>) it is reading on the tape (symbol currently under the head), tells the machine to do the following in sequence: 
    - Either erase or write a symbol (replacing a<sub>j</sub> with a<sub>j1</sub>).
    - Move the head (which is described by d<sub>k</sub> and can have values: 'L' for one step left or 'R' for one step right or 'N' for staying in the same place).
    - Assume the same or a new state as prescribed (go to state q<sub>i1</sub>).

#### Example

- Consider a simple problem (https://pythonprogramming.net/monte-carlo-simulator-python/)
    - A user can "roll" the metaphorical dice for an outcome of 1 to 100.
    - If the user rolls anything from 1-50, the "house" wins.
    - If the user rolls anything from 51 to 99, the "user" wins.
    - If the user rolls a 100, they lose.


- In this model, the house maintains a mere 1% edge, which is much smaller than the typical house edge .


- Can someone beat the odds. We will simulate a bunch of bettors. Each bettor
    - Starts with \$1000.
    - Bets \$100 on each round.
    

- What happens if we run the game for 100 rounds? How many bettors do well and how many do poorly?

```python
import random
import matplotlib
import matplotlib.pyplot as plt

# Example of a function.
def rollDice():
    
    # Example of calling a function.
    roll = random.randint(1,100)

    # Example of boolean operators inside
    # example of control flow.
    if roll == 100:
        return False
    elif roll <= 50:
        return False
    elif 100 > roll >= 50:
        return True


'''
Simple bettor, betting the same amount each time.
'''
# Example of function with input variables/objects.
def simple_bettor(funds,initial_wager,wager_count):
    
    # Example of numerical built in types.
    value = funds
    wager = initial_wager

    # Examples of built-in list types.
    # wager X
    wX = []

    #value Y
    vY = []

    # Example of assignment operator
    # change to 1, to avoid confusion so we start @ wager 1
    # instead of wager 0 and end at 100. 
    currentWager = 1

    # Example of boolean operator in control flow.
    #           change this to, less or equal.
    while currentWager <= wager_count:
        
        # Example of control flow inside control flow.
        if rollDice():
            
            # Examples of numerical and list operators, and
            # assignment operators.
            value += wager
            # append #
            wX.append(currentWager)
            vY.append(value)
            
        else:
            value -= wager
            # append #
            wX.append(currentWager)
            vY.append(value)

        currentWager += 1
        
    #print 'Funds:', value

    plt.plot(wX,vY)
    

def run_it(rounds):
    x = 0

    # start this off @ 1, then add, and increase 50 to 500, then 1000
    while x < 100:
        simple_bettor(1000,100,rounds)
        x += 1


    # Introduction to graphical plotting.
    plt.ylabel('Account Value')
    plt.xlabel('Wager Count')
    N = 2
    params = plt.gcf()
    plSize = params.get_size_inches()
    params.set_size_inches( (plSize[0]*N, plSize[1]*N) )
    plt.title(str(rounds) + '  Betting Rounds')
    plt.show()
    
    
run_it(100)
```

Hmm. Some people do well and some people do poorly, but seems about even.

What about 1,000 rounds per bettor?


```python
run_it(1000)
```

Hmm. Maybe a little bit more in favor of the house, but not obvious.

How about 10,000 rounds?


```python
run_it(10000)
```

OK. The house is now clearly doing extremely well.

What about 100,000 rounds?

```python
run_it(100000)
```

__If you have enough bettors and enough betting, the house always does very, very well even with a small advantage.__

#### Example 2: Similar, more complex example

```python
# "Pay no attention to the code behind that curtain, or def
def what(obj, callingLocals=locals()):
    for k,v in list(callingLocals.items()):
        if v is obj:
            name = k
            print("The object's name is ", name)
            print("The object's value is ", v)
            print("The object's ID is ", id(obj))
            print("The object's type is ", type(obj))
    

my_int1 = 10
my_int2 = 10
my_int3 = 8
my_int3 = my_int3 + 2

# I am going to print out some information about my_int1
print("\nTesting for my_int1")
what(my_int1)
```

What happened?
- That is not what I expected. What happened?


- We imprecisely use the term "variable" for things like my_int1.


- my_int1 is a "name." my_int1 = 10 means, "The object (thing) 10 is also known as my_int1."


- The statements my_int2 = 10 means, "The object (thing) 10 is also known as my_int2"


- The statements my_int3 = 8, my_int3 = my_int3 + 2 mean, "Take the name my_int3 away from 8 and assign to 10."


- There is a single "thing" (object, place in memory) that holds the value 10 but has three names.


- I used the _name_ my_int1 to pass the _object_ to the function _what._ The function uses a library to
    - _Get all the names that are in use_.
    - Print out information if the name refers to the object the function received as input.

- Rounding errors
    - In base 10, there is no decimal representation for irrational numbers, e.g. 1/3 or sqrt(2).
    - Some numbers are irrational in base 2, which is what a computer uses, e.g. 0.1 <br>So, (0.1 + 0.1 + 0.1) = 0.30000000000000004, and does not equal 0.3
    - This can break some fundamental laws of mathematics (algebra), e.g.
        - math.sqrt(2) * math.sqrt(2) = 2.0000000000000004 and not 2.
        -  0.3 - 0.1 - 0.1 - 0.1 = -2.7755575615628914e-17, not 0
        - (a + b) + c may or may not equal a + (b + c)
    - This is a behavior of ALL computers and programming languages.
    - In Python, you can handle with round()
        - 0.1 + 0.1 + 0.1 does not equal 0.3
        - round(0.1 + 0.1 + 0.1) DOES equal 0.3

- Remember,
    - In a computer, everything is a binary number.

    - [ASCII](https://en.wikipedia.org/wiki/ASCII) (American Standard Code for Information Interchange) used to be very common, but was "American" centric.
        - An ASCII only encodes 127 characters, and comes from the teletype days.
        - ASCII does support non-English characters.
    - IBM and some other systems used [Extended Binary Coded Decimal Interchange Code (EBCIDIC)](https://en.wikipedia.org/wiki/EBCDIC) for a long time.  
    - [UTF-8](https://en.wikipedia.org/wiki/UTF-8) is now the most common encoding.
        - UTF-8 encodes 1,112,064 characters from [Unicode](https://en.wikipedia.org/wiki/Unicode).
        - The first 127 numbers of ASCII and UTF-8 represent the same characters, that is UTF-8 is a superset of ASCII.

__Lists are fundamental of data, analytics and plotting__

```python
import numpy as np 
import matplotlib.pyplot as plt


a = np.array([1,3]) 
print("\nVector a = ", a)
b = np.array([5,2])
print("Vector b = ", b)

print("\nThe inner product of a and b is = ", np.inner(a,b))

ad = np.add(a,b)
print("\nThe sum of a + b is the vector ", ad)

ax = plt.axes()
plt.axis([0, 7, 0, 7])
l1 = ax.arrow(0,0,a[0],a[1], color='r', label="a",head_width=0.2, head_length=0.2)
l2 = ax.arrow(0,0, b[0],b[1], color='b', label="b",head_width=0.2, head_length=0.2)
l3 = ax.arrow(0,0,ad[0],ad[1], color='g', label="a+b",head_width=0.2, head_length=0.2)
l4 = ax.arrow(a[0],a[1],(ad[0]-a[0]),(ad[1]-a[1]),head_width=0.2, head_length=0.2,color='c',label="b starting from  a")
legend = ax.legend(handles=[l1,l2, l3, l4],loc='upper left')

plt.show() 
```
